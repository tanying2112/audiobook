## 1. 数据加载与预处理

加载 bounds 从各种来源（CSV、Excel、数据库）加载数据集，并执行数据清洗，包括处理缺失值、删除重复项和数据类型转换。

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# 加载示例数据集（使用 Iris 数据集作为示例）
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

print("数据集形状:", df.shape)
print("\n前5行数据:")
print(df.head())
print("\n数据类型:")
print(df.dtypes)
```

```python
# 数据清洗：处理缺失值、删除重复项

# 检查缺失值
print("缺失值统计:")
print(df.isnull().sum())

# 删除重复项
df_clean = df.drop_duplicates()
print(f"\n删除重复项后数据量: {df_clean.shape[0]} 行")

# 数据类型转换（示例：将目标变量转为分类类型）
df_clean['target'] = df_clean['target'].astype('category')
print("\n转换后的数据类型:")
print(df_clean.dtypes)
```

## 2. 探索性数据分析

分析数据分布、相关性和模式，使用描述性统计和可视化技术（如直方图、散点图和热图）。

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

# 描述性统计
print("描述性统计:")
print(df_clean.describe())

# 数据分布可视化
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

for idx, col in enumerate(df_clean.columns[:-1]):
    axes[idx].hist(df_clean[col], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
    axes[idx].set_title(f'{col} 分布')
    axes[idx].set_xlabel(col)
    axes[idx].set_ylabel('频数')

plt.tight_layout()
plt.show()
```
